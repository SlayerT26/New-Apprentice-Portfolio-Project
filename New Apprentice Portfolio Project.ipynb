```python
# Converting all columns into floating or int values

top10_2021['Revenues']=top10_2021['Revenues'].str.replace('$','').str.replace('M','').astype(float)
top10_2021['Profits']=top10_2021['Profits'].str.replace('$','').str.replace('M','').str.replace('(','').str.replace(')','').replace('-',np.nan).astype(float)
top10_2021['Employees']=top10_2021['Employees'].str.replace(',','').astype(int)

fortune2020['Revenues']=fortune2020['Revenues'].str.replace('$','').str.replace('M','').astype(float)
fortune2020['Profits']=fortune2020['Profits'].str.replace('$','').str.replace('M','').str.replace('(','').str.replace(')','').replace('-',np.nan).astype(float)
fortune2020['Employees']=fortune2020['Employees'].str.replace(',','').astype(int)

# print(top10_2021['Employees'][0])
```
